```python
###Q3. In what situations is nominal encoding preferred over one-hot encoding? Provide a practical example.

#Nominal encoding is often used interchangeably with One-Hot Encoding. However, when discussing different encoding techniques, it's important to distinguish between methods like Label Encoding (sometimes called nominal encoding in certain contexts) and One-Hot Encoding.

#Label Encoding (Nominal Encoding in Some Contexts): Assigns a unique integer to each category.
#One-Hot Encoding: Creates binary columns for each category.

```

```python
#Step 1: Import Libraries and Create Dataset
import pandas as pd

# Sample data
data = {
    'Book_ID': [101, 102, 103, 104, 105],
    'Genre': ['Fiction', 'Non-Fiction', 'Science Fiction', 'Fantasy', 'Historical'],
    'Sales': [500, 300, 250, 450, 200]
}

df = pd.DataFrame(data)

# Display the original dataset
df.head()
```

```python
#Step 2: Apply Label Encoding (Nominal Encoding)
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Apply Label Encoding to the 'Genre' column
df['Genre_Encoded'] = label_encoder.fit_transform(df['Genre'])

# Display the label-encoded dataset
df.head()
```

```python
###Q4. Suppose you have a dataset containing categorical data with 5 unique values. Which encoding technique would you use to transform this data into a format suitable for machine learning algorithms? Explain why you made this choice.
#Step 1: Import Libraries and Create Dataset
#python
#Copy code
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Sample data
data = {
    'ID': [1, 2, 3, 4, 5],
    'Category': ['A', 'B', 'C', 'D', 'E']
}

df = pd.DataFrame(data)

# Display the original dataset
df.head()

```

```python
#Option 1: One-Hot Encoding

# Initialize OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse=False)

# Apply One-Hot Encoding to the 'Category' column
one_hot_encoded = one_hot_encoder.fit_transform(df[['Category']])

# Convert to DataFrame for better visualization
one_hot_encoded_df = pd.DataFrame(one_hot_encoded, columns=one_hot_encoder.get_feature_names_out(['Category']))

# Display the one-hot encoded dataset
one_hot_encoded_df
```

```python
#Option 2: Label Encoding

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply Label Encoding to the 'Category' column
df['Category_LabelEncoded'] = label_encoder.fit_transform(df['Category'])

# Display the label-encoded dataset
df
```
